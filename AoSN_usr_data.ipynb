```python
def func_name(df):
  df = df.drop('idcn',1)
  rf = pd.DataFrame()
  tf = df
  for idus_1 in df:
    tf = tf.drop(idus_1,1)
    for idus_2 in tf:
      d_to_ap = []
      for i in range(0,len(df)):
        if np.isnan(df[idus_1][i]) == True or (df[idus_1][i]) == 0:
          x = df[idus_1][i]
        elif np.isnan(df[idus_2][i]) == True or (df[idus_2][i]) == 0:
          x = df[idus_2][i]
        else:
          x = (df[idus_1][i],df[idus_2][i])
        d_to_ap.append(x)
      rf[idus_1+', '+idus_2] = d_to_ap

    print('Покончено с ячейкой',idus_1)
  return rf
```

```python
gn = func_name(G)
```
