```r
gn <- gnr %>% transmute(
                     badges_cnt,
                     difficulty,
                     has_teacher_cnt,
                     user_city,
                     grade = case_when(between(user_grade, 2,4) ~ "1st levle",
                                  user_grade == 5 ~ "2nd levle",
                                  between(user_grade, 6,7) ~ "3rd levle",
                                  between(user_grade, 8,10) ~ "4rd levle")
)
```

Было перебрано множество переменных, представленный здесь вариант лучший по показателям кластеризации, есдинстенный плохо выделяющийся кластер - 4


```r
gn_1 <- gn[1:4]
```

```r
distance_mat <- dist(gn_1, method = 'euclidean')
```

```r
clusters <- hclust(dist(scale(gn_1)))
plot(clusters)
```

```r
clusterCut <- cutree(clusters, 4)
```

```r
table(clusterCut, gn$grade)
```

```r
library(palmerpenguins)
library(umap)
theme_set(theme_bw(18))
```

```r
G <- gnr %>% transmute(
                     IDP = X.x,
                     exercise_problem_repeat_session,
                     is_correct,
                     total_sec_taken,
                     total_attempt_cnt,
                     used_hint_cnt,
                     is_hint_used,
                     difficulty = factor(difficulty, levels = c(0:3), labels = c("Easy", "Normal", "Hard", "Unset")),
                     learning_stage = factor(learning_stage, levels = c(0:2), labels = c("Elementary", "Junior", "Senior")),
                     content_name
                     )
```

```r
G <- G %>% 
  mutate(ID=row_number())
```

```r
G_meta <- G %>%
  select(ID, IDP, content_name, difficulty, learning_stage)
```

```r
set.seed(9841)
umap_fit <- G %>%
select(where(is.numeric)) %>%
column_to_rownames("ID") %>%
scale() %>% 
umap()
```

```r
umap_G <- umap_fit$layout %>%
  as.data.frame()%>%
  rename(UMAP1="V1",
         UMAP2="V2") %>%
  mutate(ID=row_number())

umap_G <- umap_G %>%
  inner_join(G, by="ID")

umap_G %>% head()
```

```r
umap_G %>%
  ggplot(aes(x = UMAP1, 
             y = UMAP2, 
             color = is_correct, 
             shape = learning_stage))+
  geom_point()+
  labs(x = "UMAP1",
       y = "UMAP2",
      subtitle = "UMAP plot")
```
